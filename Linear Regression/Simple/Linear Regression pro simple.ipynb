```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df1=df.rename(columns={"per capita income (US$)": "percapitaincome"})
df1.head()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income")
plt.scatter(df1.year,df1.percapitaincome,color="green",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df1[["year"]],df1.percapitaincome)
```

```python
reg.predict([[2020]])
```
