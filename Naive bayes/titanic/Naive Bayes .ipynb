```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/devik/Downloads/titanic.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1.head()
```

```python
X = df1.drop("Survived",axis="columns")
X.head()
```

```python
Y=df1.Survived
Y.head()
```

```python
dummies=pd.get_dummies(X.Sex).astype(int)
dummies.head()
```

```python
X=pd.concat([dummies,X],axis="columns")
X.head()
```

```python
X=X.drop("Sex",axis="columns")
X.head()
```

```python
X.columns[X.isna().any()]
```

```python
X["Age"]=X.Age.fillna(X.Age.mean())
X.head()
```

```python
X.Age.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
```

```python
NB.fit(X_train,Y_train)
```

```python
NB.score(X_test,Y_test)
```

```python
NB.predict_proba(X_test[:10])
```

```python
NB.predict(X_test[:10])
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train,Y_train,cv=10)
```
